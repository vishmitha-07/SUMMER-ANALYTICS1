### 1. What is name of car that has the highest horsepower?  

```python
#your code here
cars_reset = cars.reset_index()
max_hp_row = cars_reset.loc[cars_reset['horsepower'].idxmax()]

car_name = max_hp_row['names']
print("Car with highest horsepower:", car_name)
```

### 2. How many cars have mpg ≥ 35?

```python
#your code here
high_mpg_cars = cars[cars['mpg'] >= 35]
print('No of  cars with mpg ≥ 35 : ' ,len(high_mpg_cars))
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here
filtered_cars = cars[(cars['horsepower'] > 100) & (cars['weight'] < 3000)]
most_common_origin = filtered_cars['origin'].mode()[0]
print('Most common origin for cars with horsepower > 100 and weight < 3000 : ',most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here
japan_cars = cars[cars['origin'] == 'japan']
mean_acceleration = round(japan_cars['acceleration'].mean(), 2)
print(f"The mean acceleration of cars from Japan is: {mean_acceleration}")
```

### 5. Which year had the highest average mpg?

```python
#your code here
avg_mpg_per_year = cars.groupby('model_year')['mpg'].mean()
best_year = avg_mpg_per_year.idxmax()
highest_avg_mpg = round(avg_mpg_per_year.max(), 2)

print(f"The year with the highest average mpg is: {best_year} with an average mpg of {highest_avg_mpg}")
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
median_mpg = cars['mpg'].median()
above_median_mpg = cars[cars['mpg'] > median_mpg].copy()
above_median_mpg['hp_to_weight_ratio'] = above_median_mpg['horsepower'] / above_median_mpg['weight']
max_ratio = above_median_mpg['hp_to_weight_ratio'].max()
best_cars = above_median_mpg[above_median_mpg['hp_to_weight_ratio'] == max_ratio]
print("Car(s) with the best horsepower to weight ratio (among cars with above-median mpg):")
for _, row in best_cars.iterrows():
    print(f"Name: {row['names']}, Horsepower: {row['horsepower']}, Weight: {row['weight']}, MPG: {row['mpg']}, Ratio: {round(row['hp_to_weight_ratio'], 5)}")
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
grouped = cars.groupby(['model_year', 'origin'])['mpg'].mean().unstack()

# Plotting
plt.figure(figsize=(10, 6))
for origin in grouped.columns:
    plt.plot(grouped.index, grouped[origin], label=origin)

# Plot customization
plt.title('Average MPG Over the Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
features = ['mpg', 'horsepower', 'weight', 'acceleration', 'origin']
data = cars[features]

# Create a PairGrid
g = sns.PairGrid(data, hue='origin', diag_sharey=False)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
g.add_legend()

plt.suptitle("Seaborn PairGrid of Car Attributes by Origin", y=1.02)
plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
name_grouped = cars.groupby('name')
multi_year_cars = name_grouped.filter(lambda x: x['model_year'].nunique() > 1)
grouped_filtered = multi_year_cars.groupby('name')
consistency_stats = grouped_filtered['mpg'].agg(['std', 'mean', 'count']).reset_index()
consistency_stats.columns = ['name', 'std_mpg', 'avg_mpg', 'count']

consistent_cars = consistency_stats[(consistency_stats['std_mpg'] < 1.0) & (consistency_stats['count'] >= 2)]

consistent_cars_sorted = consistent_cars.sort_values(by=['count', 'avg_mpg'], ascending=[False, False])

# Display result
print("Consistent Car Models (std_mpg < 1.0, at least 2 years):")
print(consistent_cars_sorted[['name', 'count', 'avg_mpg']])
```
